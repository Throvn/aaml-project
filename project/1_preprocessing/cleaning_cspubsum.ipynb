# CSPubSum Download + Cleaning

The cleaning steps are based 100% on the steps outlined in the reference paper.

```python
import pandas as pd
```

```python
df = pd.read_csv("train_CSPubSum.csv")
df = df.rename({"Abstract": "text", "Title": "title"}, axis=1)
df.drop(['FileName'], axis=1, inplace=True)
```

```python
# Since the rest of the paper assumes tokens
# we need a way to calculate the tokens as well
# SmolLM uses the Llama 3.2 tokenizer.
from transformers import AutoTokenizer
from huggingface_hub import login
login("redacted")

tokenizer = AutoTokenizer.from_pretrained(
    "HuggingFaceTB/SmolLM-135M"
)

def valid_title_length(x: str):
    length = len(tokenizer.encode(x))
    return length <= 20 and length >= 3
def valid_text_length(x: str):
    length = len(tokenizer.encode(x))
    return length >= 20


print(len(df), "samples (raw)")
df['title'].apply(valid_title_length)
df['text'].apply(valid_text_length)
print(len(df), "samples (cleaned)")
```

```python
# save to disk
df.to_csv('cleaned/train_CSPubSum.csv')
```
